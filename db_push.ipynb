```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import openpyxl

pymysql.install_as_MySQLdb()
import MySQLdb


```

```python
engine = create_engine("mysql+mysqldb://root:rootroot@localhost:3306/multi")

conn = engine.connect()


```

```python
ncs_code_info = pd.read_csv('data/NCS_code_info.csv', encoding='cp949')
ncs_code_info
```

```python
ncs_code_info.columns = ['NCS_code', 'code1', 'code2', 'code3', 'code4']
```

```python
ncs_code_info.info()
```

```python
ncs_code_info.to_sql(name='NCS_code_info', con=engine, if_exists='append', index=False)
```

```python
# 각각의 테이블 넣기
ncs_jobname = pd.read_csv('data/NCS_jobname.csv', encoding='cp949')
ncs_jobname
```

```python
ncs_jobname.info()
```

```python
ncs_jobname = ncs_jobname[['job_name', 'NCS_code']]
```

```python
emp_info = pd.read_csv('data/원티드_채용공고_DE_DBA_MLE_테이블화.csv')
emp_info
```

```python
emp_info.columns = ['company', 'job_name', 'city', 'stack', 'NCS_code', 'link']
```

```python
emp_info['no'] = [i for i in range(1, len(emp_info)+1)]
```

```python
emp_info.to_sql(name='emp_info', con=engine, if_exists='append', index=False)
```

```python
education = pd.read_csv('data/국비지원훈련과정_수정본.csv')
education
```

```python
education_center = education[['train_center', 'address', 'center_tel']]
education = education.drop(['address', 'center_tel'], axis=1)
```

```python
education['no'] = [i for i in range(1, len(education)+1)]
```

```python
education_center
```

```python
education_center.drop_duplicates(['train_center', 'address'], keep='last', inplace=True)
```

```python
education_center.loc[education_center['train_center'] == '그린컴퓨터아트학원']
```

```python
education_center.to_sql(name='education_center', con=engine, if_exists='append', index=False)
```

```python
education.to_sql(name='education', con=engine, if_exists='append', index=False)
```
